```python
from pandas import DataFrame
Class={
    "Student_Name":["Kalpana","Aswini","Abhi","Sunayani","Himaja","Shekar","Nivitha","Nelson"],
    "Hindhi":["96","98","90","92","80","90","93","91"],
    "English":["91","94","95","80","96","90","93","90"],
    "Maths":["88","87","78","89","90","91","87","86"],
    "Physics":["98","94","99","92","90","94","99","92"],
    "Total_Marks":[373,373,362,353,356,365,372,359]
       }
data=pd.DataFrame(Class)
data
```

```python
from pandas import DataFrame
data.loc[(data["Total_Marks"]<=250,"Grade")]="E"
data.loc[(data["Total_Marks"]>=250) & (data["Total_Marks"]<=300),"Grade"]="B"
data.loc[(data["Total_Marks"]>=300) & (data["Total_Marks"]<=350),"Grade"]="A"
data.loc[(data["Total_Marks"]>=351) , ("Grade")]="O"
data.loc[(data["Grade"]=="E") ,"Pass/Fail"]="F"
data.loc[(data["Grade"]=="B") | (data["Grade"]=="A") | (data["Grade"]=="O"),"Pass/Fail"]="P"
data
```
